```python
dataset_train = np.genfromtxt("train.tab", delimiter="\t")
dataset_test = np.genfromtxt("test.tab" , delimiter="\t")

maxlen= 1
embedding_dim = 64
batch_size = 32
epochs = 10
nb_word = 50000
# nb_tag = 2


x_train = dataset_train[:,3:4]
y_train= dataset_train[:,4]
x_test = dataset_test[:,3:4]
y_test = dataset_test[:,4]

X_train = sequence.pad_sequences(x_train, maxlen=maxlen)
X_test = sequence.pad_sequences(x_test, maxlen=maxlen)
Y_train = np.asarray(y_train, dtype='int32')
Y_test = np.asarray(y_test, dtype='int32')
```

```python
model = Sequential()
model.add(Embedding(nb_word, embedding_dim))
model.add(LSTM(16,activation='sigmoid', inner_activation='hard_sigmoid', dropout_W=0.1, dropout_U=0.01))


model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])


model.summary()
```

```python
model.fit(X_train, Y_train,
                batch_size=batch_size,
                nb_epoch=epochs)
```

```python
predict = []
res = model.predict(X_test)
for i in res:
    predict.append(i)
    
    #if(res[i] == 1):
        #print(Z[i])
```
